### We will convert Estimated Income numerical data to Categorical Data by binning

```python
bins = [0, 100000, 300000, float("inf")]
labels = ["Low", "Med", "High"]

df["Income Band"] = pd.cut(df["Estimated Income"], bins = bins, labels = labels, right = False)
```

```python
df["Income Band"].value_counts().plot(kind='bar')
```

### UNIVARIATE ANALYSIS

```python
#UNIVARIATE ANALYSIS
for i, predictor in enumerate (df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Loyalty Classification", "Fee Structure", "Occupation", "Properties Owned", "Risk Weighting"]]):
    plt.figure(i)
    sns.countplot(data = df, x = predictor)
```

### BIVARIATE ANALYSIS

```python
#BIVARIATE ANALYSIS wrt GenderId (Specified in hue attribute)
for i, predictor in enumerate (df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Loyalty Classification", "Fee Structure", "Occupation", "Properties Owned", "Risk Weighting"]]):
    plt.figure(i)
    sns.countplot(data = df, x = predictor, hue = "GenderId")
```

```python
#BIVARIATE ANALYSIS wrt Nationality (Specified in hue attribute)
for i, predictor in enumerate (df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Loyalty Classification", "Fee Structure", "Occupation", "Properties Owned", "Risk Weighting"]]):
    plt.figure(i)
    sns.countplot(data = df, x = predictor, hue = "Nationality")
```

```python
#BIVARIATE ANALYSIS wrt Amount of Credit Cards (Specified in hue attribute)
for i, predictor in enumerate (df[["BRId", "GenderId", "IAId", "Amount of Credit Cards", "Nationality", "Loyalty Classification", "Fee Structure", "Occupation", "Properties Owned", "Risk Weighting"]]):
    plt.figure(i)
    sns.countplot(data = df, x = predictor, hue = "Amount of Credit Cards")
```

```python
# Histplot of value counts for different occupation

for col in categorical_cols:
    if col == "Occupation":
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel("Count")
    plt.show()
```

# NUMERICAL ANALYSIS

```python
numerical_cols = ['Age', 'Estimated Income', 'Superannuation Savings', 'Credit Card Balance', 'Bank Loans', 'Bank Deposits', 'Checking Accounts', 'Saving Accounts', 'Foreign Currency Account', 'Business Lending']  
```

## Univariate Analysis and Visualization

```python
plt.figure(figsize=(30,25))
for i, col in enumerate(numerical_cols):
    plt.subplot(4,3,i+1)
    sns.histplot(df[col], kde=True)
    plt.title(col)
plt.show()
```

## Heatmaps

```python
correlation_matrix = df[numerical_cols].corr()

plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
